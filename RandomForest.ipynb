```python
import pandas as pd
import sklearn
```

```python
# data = pd.read_csv("train_data1.csv")
data = pd.read_csv("train_data.csv")
print(data.info())
```

```python
labels = data.iloc[:-1]
# labels
```

```python
from sklearn.model_selection import train_test_split

labels = data.Label

data1 = data.drop("Label", axis = 1)
X_train, X_validation, y_train, y_validation = train_test_split(data1, labels, test_size = 0.4)
################CAUTION##################################
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, criterion = "entropy", oob_score = True)

rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

prediction = rf.predict(X_validation)
print(X_validation.shape)
accuracy = accuracy_score(y_validation,prediction)

print(accuracy)
print(rf.feature_importances_)
```
